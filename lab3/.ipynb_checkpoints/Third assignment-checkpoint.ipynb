```python
import networkx as nx
from networkx.algorithms import community
import numpy as np
import matplotlib.pyplot as plt
import collections, itertools
from random import choice
#from networkx.algorithms.community.quality import modularity
#from networkx.algorithms import community
#from networkx.algorithms.community import greedy_modularity_communities
```

## 1. Load graph

```python
G = nx.davis_southern_women_graph()
```

```python
plt.figure(figsize=(10,8))
nx.draw_networkx(G,font_size=0,node_size=50)
#Print node numbers 
print("Number of nodes:",G.number_of_nodes())
#Print edges number G.number_of_edges()
print("Number of edges:",G.number_of_edges())
```

```python
# Pay off matrix
equal_po = np.eye(2)
a_po = equal_po + [[1,0],[0,0]]
q_equal = equal_po[1,1]/(equal_po[0,0]+equal_po[1,1])
q_equal
q_a = a_po[1,1]/(a_po[0,0]+a_po[1,1])
```

```python
def init_labels():
    labels={}
    for node in G.nodes():
        labels[node] = 'b'
    return labels
```

```python
pos = nx.spring_layout(G)
plt.figure(figsize=(10,8))
nx.draw_networkx(G,pos,node_color=labels.values(),font_size=10,node_size=80)
```

```python
labels = init_labels()
```

```python
bluelist = [k for k,v in labels.items() if v=='b']
newred = choice(bluelist)
print(newred)
labels[newred] = 'r'
bluelist.remove(newred)
redlist = [k for k,v in labels.items() if v=='r']

plt.figure(figsize=(10,8))
plt.title("#blue ={},#red ={}".format(len(bluelist),len(labels)-len(bluelist)))
nx.draw_networkx(G,pos,node_color=labels.values(),font_size=10,node_size=80)  
```

```python
i = 0
while True:
    initlen = list(labels.values())
    for node in bluelist:
        edgelist = []
        for pair in list(G.edges()):
            if node not in pair:
                continue
            edgelist.append(pair[1] if pair[0] == node else pair[0])
        #print(edgelist)
        num_neig = len(edgelist)
        count = 0
        for friend in edgelist:
            if labels[friend]=='r':
                count+=1
        if count/num_neig >= q_equal:
            labels[node] = 'r'
            bluelist.remove(node)
            #redlist.append(node)
    #print(initlen,list(labels.values()))
    
    plt.figure(figsize=(10,8))
    plt.title("iteration {}:#blue ={},#red ={}".format(i,len(bluelist),len(labels)-len(bluelist)))
    nx.draw_networkx(G,pos,node_color=labels.values(),font_size=10,node_size=80) 
    plt.show()
    
    if initlen == list(labels.values()):
        break
    i+=1
```
